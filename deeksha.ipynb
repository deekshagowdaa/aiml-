```
import pandas as pd
df=pd.read_csv('/content/laptop.zip')
import matplotlib.pyplot as plt
plt.plot(df['warranty'],df['graphic_card_gb'],marker='o',linestyle='-',color='pink')
plt.title("BAR PLOT")
plt.xlabel('warranty')
plt.ylabel('graphic_card_gb')
plt.show()
```
